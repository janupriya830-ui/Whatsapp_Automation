```python
import pandas as pd
import pywhatkit
import time

# Load your Excel sheet
data = pd.read_excel(r"C:\Users\Admin\OneDrive\Desktop\RR\contacts.xlsx")

# Loop through each contact and send message
for index, row in data.iterrows():
    phone = f"+{row['Ph.no']}"  # Add '+' before number
    message = row['Message']
    
    print(f"Sending message to {row['Name']} ({phone})...")
    
    # Schedule message 1 minute from current time
    current_time = time.localtime()
    hour = current_time.tm_hour
    minute = current_time.tm_min + 1

    pywhatkit.sendwhatmsg(phone, message, hour, minute, wait_time=10)
    
    print("Message scheduled ✅\n")
    time.sleep(15)  # Pause before next message
```
